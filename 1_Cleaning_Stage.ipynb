```python
import pandas as pd
import numpy as np
df = pd.read_csv("netflix_titles.csv")
print("-Original Data Info")
df.info()
print()
print(df.isnull().mean().sort_values(ascending=False))
```

```python
# In this data we have 3 rows that have their duration values sitting in the 'rating' column instead, this is a shifted column bug in the source data.
# So its not something to guess around.

mask = df['rating'].isin(['74min', '84min', '66min'])
print(f"Found{mask.sum()} rows with a duration values stuck in 'rating':")
print(df.loc[mask,['title', 'rating', 'duration']])

df.loc[mask, ' duration'] = df.loc[mask, 'rating']
df.loc[mask, 'rating'] = np.nan
```

```python
# Rating has few missing values and there is no way to recover them, we have to drop these rows
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
# Duration too have missing values and we cant recover them, i will do the same thing, drop rows
df= df.dropna(subset =['duration']).copy()
print(f"Rows after cleaning: {len(df)}")
```

```python
processed_file ="netflix_preprocessed.csv"
df.to_csv(processed_file, index = False)
print(f"Processed data save to {processed_file}")
```
