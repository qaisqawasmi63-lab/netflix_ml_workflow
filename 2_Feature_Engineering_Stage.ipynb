```python
import pandas as pd
#Loading previous preprocessed data.
df = pd.read_csv("netflix_preprocessed.csv")
```

```python
#Replacing date_added into year/month
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors = 'coerce')
df['Year_Added'] = df['date_added'].dt.year
df['Month_Added'] = df['date_added'].dt.month
df['Year_Added'] = df['Year_Added'].fillna(df['Year_Added'].median())
df['Month_Added'] = df['Month_Added'].fillna(df['Month_Added'].median())
```

```python
df['Duration_Value'] = df ['duration'].str.extract(r'(\d+)').astype(float)
df['Duration_Unit'] = df ['duration'].str.extract(r'([A-Za-z]+)')
```

```python
df['Has_Director'] = df['director'].notna().astype(int)
df['Has_Country'] = df['country'].notna().astype(int)
df['Cast_Count'] = df['cast'].fillna('').apply(lambda x: len([c for c in x.split(',') if c.strip()]))
df['Primary_Country'] = df['country'].fillna('Unknown').apply(lambda x: x.split(',')[0].strip())
```

```python
# Listing (genre) - count + primary genre
df['Genre_Count'] = df['listed_in'].apply(lambda x: len([g for g in x.split(',')[0].strip()]))
df['Primary_Genre'] = df['listed_in'].apply(lambda x: x.split(',')[0].strip())
```

```python
# Cheap text signals from title and description, without full NLP
df['Title_Length'] = df['title'].str.len()
df['Description_WordCount'] = df['description'].str.split().apply(len)
```

```python
# We are going to collapse rare categories into 'Other' so one-hot encoding doesnt explode into hundreds of columns.
top_countries = df ['Primary_Country'].value_counts().nlargest(10).index
df['Primary_Country'] = df['Primary_Country'].where(df['Primary_Country'].isin(top_countries), 'Other')

top_genres =df['Primary_Genre'].value_counts().nlargest(10).index
df['Primary_Genre'] = df['Primary_Genre'].where(df['Primary_Genre'].isin(top_genres), 'Other')
```

```python
engineered_file = "netflix_engineered.csv"
df.to_csv(engineered_file, index = False)
print(f"Data saved to {engineered_file}")
```
